```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# data load
df = pd.read_excel("../data/data1.xlsx")
```

```python
# rolling retention 구하기
# sign_up month만 가져오기
df["sign_up_month"] = df["sign_up"].dt.month
df["last_login_month"] = df["last_login"].dt.month
```

```python
# pivot table
retention_df = pd.pivot_table(data=df, index="sign_up_month",
               columns="last_login_month", values="user_id", aggfunc = "count")
```

```python
retention_df["합계"] = retention_df.sum(axis = 1)
retention_df.fillna(0, inplace = True)
```

```python
# 월별 가입자
# pandas heatmap
retention_df.style.background_gradient(cmap='summer')
```

```python
for idx in range(len(retention_df)):
    num = retention_df.iloc[idx, -1]
    for month in range(idx +1, 13):
        month_num = retention_df.loc[idx + 1, month]
        retention_df.loc[idx + 1, month] = num
        num -= month_num
```
